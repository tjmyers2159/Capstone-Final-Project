```python
import pandas as pd

df = pd.read_csv('survey_results_public.csv')
df
```

```python
df_usa = df[df['Country'] == 'United States of America']
df_usa
```

```python
df_usa_missing = df_usa.dropna(subset=['ConvertedCompYearly'])
df_usa_missing
```

```python
df_usa_employed = df_usa_missing[df_usa_missing['Employment'] == 'Employed']
df_usa_employed
```

```python
df_usa_comp_filtered = df_usa_employed[(df_usa_employed['ConvertedCompYearly'] >= 30000) & (df_usa_employed['ConvertedCompYearly'] <= 500000)]
df_usa_comp_filtered
```

```python
df_usa_years_filtered = df_usa_comp_filtered[(df_usa_comp_filtered['YearsCode'] >= 1) & (df_usa_comp_filtered['YearsCode'] <= 30)]
df_usa_years_filtered
```

```python
selected_columns = ['LanguageHaveWorkedWith', 'ConvertedCompYearly', 'YearsCode', 'ResponseId']
df_final = df_usa_years_filtered[selected_columns]
df_final
```

```python
df_final.to_csv('survey_USA_cleaned.csv', index=False)
```
